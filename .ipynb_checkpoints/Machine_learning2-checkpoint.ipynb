```python
df = data.copy()
df
```

```python
# 행 삭제 (위에 조건에 해당 하는 인덱스 명)
data.drop(data[cond].index, axis=0,inplace=True)
data
```

```python
# apply 함수 예시 (칼로리가 1300 보다 큰 값과 작은 값)
def cal(x):
    if x >= 1300:
        return "yes"
    else:
        return "no"

df['칼로리'].apply(cal)
```

```python
# apply 적용해서 새로운 컬럼 생성 (칼로리 칼럼 활용)
df['살찔까요'] = df['칼로리'].apply(cal)
df
```

```python
# lambda, apply 활용 : 새로운 컬럼 추가하는 다른 방법
df['고민'] = df['칼로리'].apply(lambda x: '먹지말자' if x >= 1300 else '먹자')
df
```
