```python
# csv 파일 불러오기
data = pd.read_csv('final_modudak.csv')
data
```

```python
# 데이터 변경
data.loc[2,'원산지'] = '미국'
```

```python
# 활용할 데이터 선택
df = data[['가격','호수','칼로리','원산지','살찔까요']].copy()
df
```

```python
# 데이터 전처리 하기!!!
# 1. 범주형 데이터
# type을 숫자형으로 바꿔주어야 한다 (레이블, 원핫 인코딩)
df.info()
```

```python
# 살찔까요 레이블인코딩
df['살찔까요'] = le.fit_transform(df['살찔까요'])
df
```

```python
# 여러개의 컬럼을 한번에 인코딩 하기
# 다시 데이터프레임 갖고오기
df = data[['가격','호수','칼로리','원산지','살찔까요']].copy()
df
```

```python
# object 타입 선택 원산지, 살찔까요
# cols = ['원산지','살찔까요']
# 컬럼이 많을 때 컬럼 명 확인하는 법
df.select_dtypes(include='object').columns
```

```python
# 한 번에 레이블 인코딩

for col in cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
```

```python
# 인코딩 결과 확인
df
```

```python
# 원-핫 인코딩
# 활용할 데이터 다시 선택(초기화)
df = data[['가격','호수','칼로리','원산지','살찔까요']].copy()
df
```

```python
# 원-핫 인코딩(sparse인자 넣어야함)
one = OneHotEncoder(sparse=False) #기본값 True, False:array형
cat = one.fit_transform(df[['원산지']])
cat
```

```python
# 카테고리 화 (원핫코딩은 인코딩 값에 따라 컬럼 종류가 많아짐)
one.categories_
```

```python
# 피처(컬럼) 이름과 카테고리, 컬럼명 지정 !! 주로 원핫인코딩 컬럼 명은 get_feature로
one.get_feature_names_out()
```

```python
# 데이터프레임으로 변환
df_cat = pd.DataFrame(cat, columns=one.get_feature_names_out())
df_cat
```

```python
# 데이터 프레임 합치기 (데이터 + 원핫 인코딩 데이터)
df = pd.concat([df,df_cat],axis=1)
```

```python
# 기존 컬럼 삭제
df = df.drop(['원산지'],axis=1)
df
```

```python
# 원핫 인코딩(심화) : 여러 개 컬럼 한번에 인코딩 하기
# 데이터 초기화
df = data[['가격','호수','칼로리','원산지','살찔까요']].copy()
df
```

```python
# 원 핫 인코딩
cols = df.select_dtypes(include='object').columns

ohe = OneHotEncoder(sparse=False) # 기본값 True, False:array형
cat = ohe.fit_transform(df[cols])

df_cat = pd.DataFrame(cat, columns=ohe.get_feature_names_out())
df_cat
```

```python
# 데이터프레임 합치기
df = pd.concat([df,df_cat],axis=1)
df.head()
```

```python
# 컬럼 삭제
df = df.drop(cols, axis=1)
df
```

```python
# 원핫인코딩 : 판다스 활용 (!꿀팁)
# 데이터 초기화
df = data[['가격','호수','칼로리','원산지','살찔까요']].copy()
df
```

```python
# 판다스를 활용한 원핫 인코딩 : 판다스를 사용한 인코딩은 컬럼 수가 다를 수가 있다
df = pd.get_dummies(df)
```

```python
# 전처리 : 수치형 데이터
# 표준화 : 평균이 0, 분산이 1 인 표준정규분포로 변환
# 원-핫 인코딩 데이터 호출
df.head()
```

```python
# StandardScaler # 모델에 최적화 시켜서 넣기 위해 표준화
scaler = StandardScaler()
scaler.fit_transform(df[['가격']]) # 데이터 프레임 형으로 대괄호 2개
```

```python
# StandardScaler 여러 변수에 적용
cols = ['가격','호수','칼로리']

scaler = StandardScaler()
df[cols] = scaler.fit_transform(df[cols])
df
```

```python
# 민맥스스케일러(MinMaxScaler) : 0~1 사이 값으로 변환 (데이터 사이즈 축소)
# MinMaxScaler
cols = ['가격','호수','칼로리']
scaler = MinMaxScaler()
df[cols] = scaler.fit_transform(df[cols])
df
```
